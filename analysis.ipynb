# 🚦 City Traffic Peak Hours Analysis
### Analyzing Uber NYC Pickup Data to identify peak hours, daily patterns, and demand hotspots

**Dataset:** Uber Pickups in New York City (fivethirtyeight, via Kaggle)

**Goal:** Identify peak traffic hours, weekday vs weekend patterns, and geographic demand hotspots to help understand urban transport bottlenecks.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load Data

Download the dataset from Kaggle (`fivethirtyeight/uber-pickups-in-new-york-city`) and place the CSV files inside the `data/` folder.

We'll start with a single month file (e.g. `uber-raw-data-apr14.csv`). You can later combine multiple months using `pd.concat`.

```python
df = pd.read_csv('../data/uber-raw-data-apr14.csv')
print(df.shape)
df.head()
```

## 3. Data Cleaning & Feature Engineering

```python
# Convert Date/Time column to datetime
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%m/%d/%Y %H:%M:%S')

# Extract useful features
df['Hour'] = df['Date/Time'].dt.hour
df['Day'] = df['Date/Time'].dt.day
df['Weekday'] = df['Date/Time'].dt.day_name()
df['WeekdayNum'] = df['Date/Time'].dt.dayofweek  # 0=Monday
df['Month'] = df['Date/Time'].dt.month_name()

# Check for nulls/duplicates
print('Nulls:\n', df.isnull().sum())
print('Duplicates:', df.duplicated().sum())

df.drop_duplicates(inplace=True)
df.head()
```

## 4. Peak Hour Analysis

```python
hourly_counts = df.groupby('Hour').size()

plt.figure(figsize=(12,6))
sns.barplot(x=hourly_counts.index, y=hourly_counts.values, palette='viridis')
plt.title('Total Pickups by Hour of Day', fontsize=14)
plt.xlabel('Hour of Day')
plt.ylabel('Number of Pickups')
plt.savefig('../images/hourly_pickups.png', dpi=150, bbox_inches='tight')
plt.show()

peak_hour = hourly_counts.idxmax()
print(f'Peak Hour: {peak_hour}:00 with {hourly_counts.max()} pickups')
```

## 5. Weekday vs Weekend Patterns

```python
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_counts = df.groupby('Weekday').size().reindex(weekday_order)

plt.figure(figsize=(12,6))
sns.barplot(x=weekday_counts.index, y=weekday_counts.values, palette='magma')
plt.title('Total Pickups by Day of Week', fontsize=14)
plt.xlabel('Day')
plt.ylabel('Number of Pickups')
plt.savefig('../images/weekday_pickups.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Hour vs Weekday Heatmap (Key Visualization)

```python
pivot = df.pivot_table(index='Weekday', columns='Hour', values='Date/Time', aggfunc='count').reindex(weekday_order)

plt.figure(figsize=(16,7))
sns.heatmap(pivot, cmap='YlOrRd', linewidths=0.3, annot=False)
plt.title('Pickup Density: Hour of Day vs Day of Week', fontsize=14)
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.savefig('../images/heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Geographic Hotspots (Lat/Lon Density)

```python
plt.figure(figsize=(10,10))
plt.scatter(df['Lon'], df['Lat'], s=0.5, alpha=0.2, color='steelblue')
plt.title('Pickup Location Density (NYC)', fontsize=14)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig('../images/geo_density.png', dpi=150, bbox_inches='tight')
plt.show()
```
